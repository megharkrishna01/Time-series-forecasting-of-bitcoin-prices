```python
import numpy as np
import pandas as pd
data = np.random.randn(4, 4)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])
print(df)
```

```python
df = df.rename(columns={'A': 'Random value 1', 'B': 'Random value 2' , 'C': 'Random value 3' , 'D': 'Random value 4'})
print(df)
```

```python
desp_stats =df.describe()
print(desp_stats)
```

```python
null_values = df.isnull()
col_datatypes = df.dtypes
print(null_values)
print(col_datatypes)
```

```python
loc_method = df.loc[:, ['Random value 2', 'Random value 3']]
print(loc_method)
```

```python
ind_locmethod = df.iloc[:, [1, 2]]
print(ind_locmethod)
```
